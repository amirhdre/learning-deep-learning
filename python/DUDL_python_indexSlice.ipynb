# VIDEO: Slicing, part 1

```python
alist = list( range(5,11) )
alist
```

```python
print( alist[0:2] )
print( alist[0:1] )
```

```python
alist[2:5]
```

```python
alist[2:]
```

```python
alist[:4]
```

```python
alist[::2]
```

```python
# using variables
start = 1
stop  = 4
skip  = 2

alist[start:stop:skip]
```

```python
# to reverse a list
alist[::-1]
```
