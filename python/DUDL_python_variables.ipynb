# Printing mixed variable types

```
age = 26

print(firstName + ' is ' + str(age) + ' years old.')
```

```
# using string interpolation
print('%s is %g years old.' %(firstName,age))
```

```
# using f-strings
print(f'{firstName} is {age} years old.')
```
